```python
import cv2
import os

# Directory containing the videos
video_dir = 'G:\\IITB dataset\\IEMOCAP_full_release\\Session1\\dialog\\avi\\DivX'
# Directory to save the extracted frames
output_dir = 'G:\\IITB dataset\\IEMOCAP_full_release\\Session1\\dialog\\avi\\new'

os.makedirs(output_dir, exist_ok=True)

def extract_frames(video_path, output_dir, frame_rate=1):
    cap = cv2.VideoCapture(video_path)
    video_name = os.path.basename(video_path).split('.')[0]
    frame_count = 0
    
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        if frame_count % frame_rate == 0:
            frame_name = f"{video_name}_frame_{frame_count}.jpg"
            frame_path = os.path.join(output_dir, frame_name)
            cv2.imwrite(frame_path, frame)
        frame_count += 1
    
    cap.release()

# Extract frames from all videos in the directory
for video_file in os.listdir(video_dir):
    video_path = os.path.join(video_dir, video_file)
    extract_frames(video_path, output_dir)
```

```python
from fer import FER
import cv2
import os
import pandas as pd

# Directory containing the frames
frame_dir = 'G:\\IITB dataset\\IEMOCAP_full_release\\Session1\\dialog\\avi\\new'
# Directory to save the detected faces
output_face_dir = 'G:\\IITB dataset\\IEMOCAP_full_release\\Session1\\dialog\\avi\\NEW 2'
os.makedirs(output_face_dir, exist_ok=True)

# Initialize the FER model
detector = FER()

# DataFrame to store the results
results = []

def detect_emotion(frame_path):
    img = cv2.imread(frame_path)
    emotions = detector.detect_emotions(img)
    if emotions:
        for emotion in emotions:
            results.append({
                'frame': frame_path,
                'actor': os.path.basename(frame_path).split('_')[0],
                'emotion': max(emotion['emotions'], key=emotion['emotions'].get),
                'score': emotion['emotions'][max(emotion['emotions'], key=emotion['emotions'].get)]
            })
            face = emotion['box']
            x, y, w, h = face
            face_img = img[y:y+h, x:x+w]
            face_img_path = os.path.join(output_face_dir, os.path.basename(frame_path))
            cv2.imwrite(face_img_path, face_img)

# Detect emotions in all frames
for frame_file in os.listdir(frame_dir):
    frame_path = os.path.join(frame_dir, frame_file)
    detect_emotion(frame_path)

# Save results to a CSV file
df = pd.DataFrame(results)
df.to_csv('emotion_results.csv', index=False)
```

```python
import pandas as pd
df=pd.read_csv("emotion_results.csv")
```

```python
# Assign ratings based on emotion
emotion_to_rating = {
    'happy': 5,
    'sad': 1,
    'neutral': 3,
    'angry': 2,
    'surprised': 4,
    'disgust': 2,
    'fearful': 1,
    'excited': 4
}

df['rating'] = df['emotion'].map(emotion_to_rating)
df.to_csv('rated_emotion_results.csv', index=False)
```
